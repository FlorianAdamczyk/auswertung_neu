```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
def dat2array(file_path):
    data=[]
    with open(file_path, 'r', encoding='utf-8') as file: 
        
        for i,line in enumerate(file):
              a=line.replace(',','.').replace('\t', ';').replace('\n','')

              if i >= 39: data.append(a.split(';'))

    freqs = [float(val[0]) for val in data] 
    intensities = [float(val[1]) for val in data]

    return freqs,intensities
    
```

```python
fig, ax = plt.subplots()
freqs, intensities = dat2array('TCBC_1200nm_75uW_20cmLens_291K_SHG.asc')
ax.plot(freqs, intensities, label='291K, 1200nm, 75uW')
ax.set_xlim(500, 700)
ax.set_ylim(0, 2000)
ax.legend(loc='upper right')
ax.set_xlabel('Wavelength in nm')
ax.set_ylabel('SHG intensity in a.u.')
fig.suptitle('SHG measurement of TCBC at 1200nm and 75W (291K)', fontsize=16)
plt.savefig('./TCBC_1200nm_75uW_20cmLens_291K_SHG.png', dpi=600)
plt.show()
```

```python
# Lade und plote das THG-Signal
freqs_thg, intensities_thg = dat2array('TCBC_1200nm_433uW_20cmLens_291K_THG.asc')
fig2, ax2 = plt.subplots()
ax2.plot(freqs_thg, intensities_thg, color='red', label='291K, 1200nm, 433uW')
ax2.set_xlim(320, 500)
ax2.set_ylim(1000, 4000)  # ggf. anzupassende Achsengrenzen
ax2.legend(loc='upper right')
ax2.set_xlabel('Wavelength in nm')
ax2.set_ylabel('THG intensity in a.u.')
fig2.suptitle('THG measurement of TCBC at 1200nm and 433W (291K)', fontsize=16)
plt.savefig('./TCBC_1200nm_433uW_20cmLens_291K_THG.png', dpi=600)
plt.show()
```

```python
# Combine SHG and THG plots into one diagram
fig, ax = plt.subplots()
freqs, intensities = dat2array('TCBC_1200nm_75uW_20cmLens_291K_SHG.asc')
freqs_thg, intensities_thg = dat2array('TCBC_1200nm_433uW_20cmLens_291K_THG.asc')

# Plot SHG data
ax.plot(freqs, intensities, label='SHG: 75W, slit width: 110m', color='blue')

# Plot THG data
ax.plot(freqs_thg, intensities_thg, label='THG: 433W, slit width: 330m', color='red')

# Set x and y limits to accommodate both datasets
ax.set_xlim(300, 700)
ax.set_ylim(0, 4000)

# Add legend, labels, and title
ax.legend(loc='upper right')
ax.set_xlabel('Wavelength in nm')
ax.set_ylabel('Intensity in a.u.')
fig.suptitle('SHG and THG Measurements of TCBC at 1200nm (291K)', fontsize=16)

# Save and display the plot
plt.savefig('./TCBC_1200nm_combined_SHG_THG.png', dpi=600)
plt.savefig('./TCBC_1200nm_combined_SHG_THG.pdf', dpi=600)
plt.show()
```
